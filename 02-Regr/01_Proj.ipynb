```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/alexeygrigorev/datasets/master/car_fuel_efficiency.csv")
df.head()
```

```python
colz = list(df.columns)
colz = [col.lower().replace(' ', '_') for col in colz]

df.columns = colz
df.head()
```

```python
df.dtypes
```

```python
string_colz = list(df.dtypes[df.dtypes== 'object'].index)
for col in string_colz:
    df[col] = df[col].str.lower().str.replace(' ','_').str.replace('-','_')

df.head()
```

```python
for col in  df.columns:
    print('\n %%%%%%%%%%%%%%%%%%%%%%%%%%%')
    print(col)
    if (df[col].dtypes == 'object') | (col in ['num_doors','model_year']):
        print(df[col].value_counts())
    else:
        df[col].describe()
```
